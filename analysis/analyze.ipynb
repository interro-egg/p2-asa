```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="whitegrid", palette="colorblind") # https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme
df = pd.read_csv("./1643324046.csv") # cvs com headers: V, E, mean, ...
```

```python
df["V+E"] = df["V"] + df["E"]
```

```python
df
```

```python
#plt.figure(figsize=(20, 8))
ax = sns.lineplot(x="V+E", y="mean", data=df, err_style="bars", ci="sd")
ax.set_xlabel("V+E")
ax.set_ylabel("t")
ax.set_xscale("function", functions=(lambda x: x, lambda x: x))
plt.title("Tempo de Execução")
plt.savefig("./graph.png")
```
